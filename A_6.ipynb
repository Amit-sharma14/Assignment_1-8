```python
# Step 1. Import the necessary libraries

import pandas as pd
import numpy as np
```

```python
# Step 2. Import the dataset

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
```

```python
# Step 3. Assign it to a variable called wine

wine = df
wine.head()
```

```python
# Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

wine.drop(wine.columns[[0, 3, 6, 8, 10, 12, 13]], axis=1)
```

```python
# Step 5. Assign the columns as below:
#The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):
# 1) Alcohol
# 2) Malic acid
# 3) Ash
# 4) Alcalinity of ash
# 5) Magnesium
# 6) Total phenols
# 7) Flavanoids
# 8) Nonflavanoid phenols
# 9) Proanthocyanins
# 10)Color intensity
# 11)Hue
# 12)OD280/OD315 of diluted wines
# 13)Proline

wine.columns =['type','Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
wine.head()
```

```python
# Step 6. Set the values of the first 3 rows from alcohol as NaN

wine.loc[0:2, ['Alcohol']] = [np.nan]
wine
```

```python
# Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

wine.loc[2:3, ['Magnesium']] = [np.nan]
wine.head()
```

```python
# Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

wine['Alcohol'].replace(np.nan,10,inplace=True)
wine['Magnesium'].replace(np.nan,100,inplace=True)
wine
```

```python
# Step 9. Count the number of missing values

wine.isnull().sum() # no missing values
```

```python
# Step 10. Create an array of 10 random numbers up until 10


randnums= np.random.randint(1,10,10)
randnums
```

```python
# Step 11. Use random numbers you generated as an index and assign NaN value to each of cell.

wine.loc[randnums] = np.nan
wine.head(10)
```

```python
# Step 12. How many missing values do we have?

wine.isnull().sum()
```

```python
# Step 13. Delete the rows that contain missing values

wine.dropna(inplace=True)
wine
```

```python
# Step 14. Print only the non-null values in alcohol

wine.loc[wine['Alcohol'].notnull()]
```

```python
# Step 15. Reset the index, so it starts with 0 again

wine.reset_index(inplace = True, drop = True)
wine
```
